```python
# Example 1. `range(10)` - 0부터 9까지 숫자 만들기
# range(10)은 0부터 9까지의 숫자 범위를 만든다.
# stop 값인 10은 포함되지 않는다.
x = range(10)

print(x)        # range 객체 자체 출력
print(list(x))  # 실제 숫자를 보기 위해 list로 변환
```

```python
# Example 2. `range(3, 10)` - 시작값과 끝값 지정하기
# range(3, 10)은 3부터 9까지의 숫자 범위를 만든다.
# stop 값인 10은 포함되지 않는다.
x = range(3, 10)

print(x)
print(list(x))
```

```python
# Example 3. `range(3, 10, 2)` - 간격 지정하기
# range(3, 10, 2)는 3부터 9까지 2씩 증가한다.
x = range(3, 10, 2)

print(x)
print(list(x))
```

```python
# Example 4. `range()`를 `for`문에 사용하기
# range(10)을 for문에 넣으면 0부터 9까지 하나씩 꺼내서 반복한다.
for i in range(10):
    print(i)
```

```python
# Example 5. `range()` 결과를 리스트로 보기
# range 객체를 list로 바꾸면 실제 숫자들이 보인다.
print(list(range(5)))        # 0부터 4까지
print(list(range(1, 6)))     # 1부터 5까지
print(list(range(5, 20, 3))) # 5부터 19까지 3씩 증가
```

```python
# Example 6. `range` 인덱싱과 슬라이싱
# range도 순서가 있는 자료형이므로 인덱싱과 슬라이싱이 가능하다.
r = range(10)

print(r[2])        # 인덱스 2 위치의 값: 2
print(r[:3])       # 0번부터 2번까지의 range 객체
print(list(r[:3])) # 보기 쉽게 list로 변환
```

```python
# Example 7. `in`으로 값이 포함되어 있는지 확인하기
# range(0, 10, 2)는 0, 2, 4, 6, 8을 만든다.
r = range(0, 10, 2)

print(6 in r)  # 6은 있으므로 True
print(7 in r)  # 7은 없으므로 False
```

```python
# Example 8. `len()`으로 range의 길이 구하기
# range(0, 10, 2)는 [0, 2, 4, 6, 8]이므로 길이는 5이다.
r = range(0, 10, 2)
print(len(r))
```

```python
# Example 12. 리스트 길이 구하기
# 리스트에 들어 있는 항목의 개수를 구한다.
cars = ["Ford", "Volvo", "BMW"]
x = len(cars)

print(x)
```

```python
# Example 14. 리스트 끝에 값 추가하기
# append()는 리스트의 마지막에 값을 추가한다.
cars = ["Ford", "Volvo", "BMW"]
cars.append("Honda")

print(cars)
```

```python
# Example 15. `pop()`으로 특정 위치의 값 삭제하기
# pop(1)은 인덱스 1 위치의 값을 삭제한다.
cars = ["Ford", "Volvo", "BMW"]
removed_car = cars.pop(1)

print("삭제된 값:", removed_car)
print("남은 리스트:", cars)
```

```python
# Example 16. `remove()`로 특정 값을 삭제하기
# remove()는 값 자체를 기준으로 삭제한다.
cars = ["Ford", "Volvo", "BMW", "Volvo"]
cars.remove("Volvo")

print(cars)  # 첫 번째 Volvo만 삭제된다.
```

```python
# 추가 실습. 리스트 메서드 한 번에 확인하기
# PDF에 나온 리스트/배열 메서드들을 간단히 실습한다.
nums = [3, 1, 2, 2]
print("처음 리스트:", nums)

nums.append(4)
print("append(4):", nums)

print("count(2):", nums.count(2))
print("index(3):", nums.index(3))

nums.insert(1, 10)
print("insert(1, 10):", nums)

nums.extend([5, 6])
print("extend([5, 6]):", nums)

nums.reverse()
print("reverse():", nums)

nums.sort()
print("sort():", nums)

copied_nums = nums.copy()
print("copy():", copied_nums)

nums.clear()
print("clear():", nums)
```

```python
# Example 17. `mymodule.py` 파일 만들기
# mymodule.py 파일을 만들어 모듈을 준비한다.
# %%writefile은 주피터 노트북에서 파일을 저장할 때 사용하는 명령어이다.

module_code = """
def greeting(name):
    print("Hello, " + name)
"""

with open("mymodule.py", "w", encoding="utf-8") as f:
    f.write(module_code)

print("mymodule.py 파일 생성 완료")
```

```python
# Example 18. 만든 모듈 불러와서 함수 사용하기
# 모듈을 import한 뒤, 모듈 이름.함수 이름 형태로 사용한다.
import importlib
import mymodule

importlib.reload(mymodule)  # 파일을 수정했을 때 최신 내용으로 다시 불러오기
mymodule.greeting("Jonathan")
```

```python
# Example 19. 모듈 안에 변수와 딕셔너리 저장하기
# mymodule.py에 함수와 딕셔너리 변수를 함께 저장한다.
module_code = """
def greeting(name):
    print("Hello, " + name)

person1 = {
    "name": "John",
    "age": 36,
    "country": "Norway"
}
"""

with open("mymodule.py", "w", encoding="utf-8") as f:
    f.write(module_code)

print("mymodule.py에 person1 딕셔너리 추가 완료")
```

```python
# Example 20. 모듈 안의 딕셔너리 값 사용하기
# 모듈 안의 변수는 모듈이름.변수이름 형태로 접근한다.
import importlib
import mymodule

importlib.reload(mymodule)
a = mymodule.person1["age"]
print(a)
```

```python
# Example 21. 모듈 이름을 짧게 바꿔 사용하기
# mymodule을 mx라는 별명으로 불러온다.
import importlib
import mymodule as mx

importlib.reload(mx)
a = mx.person1["age"]
print(a)
```

```python
# 추가 실습. 자주 쓰는 `strftime()` 형식 코드 확인하기
# strftime()의 여러 형식 코드를 확인한다.
import datetime

x = datetime.datetime(2018, 12, 31, 17, 41, 8, 548513)

print("%Y-%m-%d:", x.strftime("%Y-%m-%d"))  # 연-월-일
print("%H:%M:%S:", x.strftime("%H:%M:%S"))  # 시:분:초
print("%A:", x.strftime("%A"))              # 요일 전체 이름
print("%B:", x.strftime("%B"))              # 월 전체 이름
print("%c:", x.strftime("%c"))              # 지역 형식의 날짜와 시간
```

```python
# Example 36. JSON 문자열을 파이썬 딕셔너리로 변환하기
# JSON 문자열을 파이썬 딕셔너리로 변환한다.
import json

# JSON 형식의 문자열
x = '{ "name":"John", "age":30, "city":"New York"}'

# json.loads()를 사용하면 문자열이 딕셔너리로 바뀐다.
y = json.loads(x)

print(y)
print(y["age"])
```

```python
# Example 37. 파이썬 딕셔너리를 JSON 문자열로 변환하기
# 파이썬 딕셔너리를 JSON 문자열로 변환한다.
import json

x = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

y = json.dumps(x)

print(y)
print(type(y))  # JSON 문자열이므로 str 타입이다.
```

```python
# Example 38. 여러 파이썬 자료형을 JSON 문자열로 바꾸기
# 다양한 파이썬 자료형을 JSON 문자열로 변환한다.
import json

print(json.dumps({"name": "John", "age": 30}))
print(json.dumps(["apple", "bananas"]))
print(json.dumps(("apple", "bananas")))
print(json.dumps("hello"))
print(json.dumps(42))
print(json.dumps(31.76))
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

```python
# Example 39. 여러 자료형이 들어 있는 객체를 JSON으로 변환하기
# 여러 자료형이 섞인 파이썬 객체를 JSON 문자열로 변환한다.
import json

x = {
    "name": "John",
    "age": 30,
    "married": True,
    "divorced": False,
    "children": ("Ann", "Billy"),
    "pets": None,
    "cars": [
        {"model": "BMW 230", "mpg": 27.5},
        {"model": "Ford Edge", "mpg": 24.1}
    ]
}

print(json.dumps(x))
```

```python
# Example 40. `indent`로 JSON을 보기 좋게 출력하기
# indent 옵션을 사용하면 JSON 문자열을 보기 좋게 정리할 수 있다.
import json

x = {
    "name": "John",
    "age": 30,
    "married": True,
    "divorced": False,
    "children": ("Ann", "Billy"),
    "pets": None,
    "cars": [
        {"model": "BMW 230", "mpg": 27.5},
        {"model": "Ford Edge", "mpg": 24.1}
    ]
}

print(json.dumps(x, indent=4))
```

```python
# Example 41. `separators`로 구분 기호 바꾸기
# separators 옵션으로 JSON 문자열의 구분 기호를 바꿀 수 있다.
import json

x = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

print(json.dumps(x, indent=4, separators=(". ", " = ")))
```

```python
# Example 42. `sort_keys`로 key 정렬하기
# sort_keys=True를 사용하면 key가 알파벳 순서로 정렬된다.
import json

x = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

print(json.dumps(x, indent=4, sort_keys=True))
```

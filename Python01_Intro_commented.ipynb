```python
# 예제 설명: print()를 사용해서 화면에 문자를 출력하는 가장 기본 예제입니다.
print("Hello, World!")
```

```python
# 예제 설명: if 조건문은 조건이 참일 때 들여쓰기 된 코드를 실행한다는 것을 보여줍니다.
if 5 > 2: 
    print("Five is greater than two!")
```

```python
# 예제 설명: if문 아래 코드는 반드시 들여쓰기해야 하며, 들여쓰기가 없으면 오류가 난다는 예제입니다.
if 5 > 2: 
print("Five is greater than two!")
```

```python
# 예제 설명: 각 if문 아래에 들여쓰기 된 코드가 각각 실행된다는 것을 보여줍니다.
if 5 > 2: 
    print("Five is greater than two!") 
if 5 > 2: 
        print("Five is greater than two!")
```

```python
# 예제 설명: 같은 코드 블록 안에서 들여쓰기가 맞지 않으면 오류가 난다는 예제입니다.
if 5 > 2: 
    print("Five is greater than two!") 
        print("Five is greater than two!")
```

```python
# 예제 설명: 함수 안에서 바깥쪽 전역 변수를 읽어 사용할 수 있다는 예제입니다.
x = "awesome" 
def myfunc(): 
    print("Python is " + x) 
myfunc()
```

```python
# 예제 설명: 함수 안의 지역 변수는 함수 밖의 전역 변수와 별개라는 것을 보여줍니다.
x = "awesome" 
def myfunc(): 
    x = "fantastic" 
    print("Python is " + x) 
myfunc() 
print("Python is " + x)
```

```python
# 예제 설명: global 키워드로 함수 안에서 전역 변수를 만들 수 있다는 예제입니다.
def myfunc(): 
    global x 
    x = "fantastic" 
myfunc() 
print("Python is " + x)
```

```python
# 예제 설명: global 키워드로 함수 안에서 기존 전역 변수 값을 바꿀 수 있다는 예제입니다.
x = "awesome" 
def myfunc(): 
    global x 
    x = "fantastic" 
myfunc() 
print("Python is " + x)
```

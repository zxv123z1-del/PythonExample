```python
with open("demofile.txt", "w", encoding="utf-8") as f:
    f.write("Hello! Welcome to demofile.txt\n")
    f.write("This file is for testing purposes.\n")
    f.write("Good Luck!")

print("demofile.txt 파일 생성 완료")
```

```python
import os

if os.path.exists("myfile.txt"):
    os.remove("myfile.txt")

f = open("myfile.txt", "x", encoding="utf-8")
f.close()

print("myfile.txt 파일이 생성되었습니다.")
print("현재 존재 여부:", os.path.exists("myfile.txt"))
```

```python
import os

if os.path.exists("myfile.txt"):
    os.remove("myfile.txt")
    print("myfile.txt 파일을 삭제했습니다.")
else:
    print("삭제할 파일이 없습니다.")
```

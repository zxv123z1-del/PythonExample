```python
# 1. Python RegEx 기본
# Example 1
# 목적: 문자열이 "The"로 시작하고 "Spain"으로 끝나는지 확인하는 예제입니다.
# ^The  : 문자열이 The로 시작해야 함
# .*    : 중간에 어떤 글자가 0개 이상 와도 됨
# Spain$: 문자열이 Spain으로 끝나야 함

import re

txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt)

print(x)
print("조건을 만족하나요?", x is not None)
```

```python
# 3. Metacharacters, 메타문자
# Example 2
# 목적: 정규표현식의 주요 메타문자가 어떤 식으로 동작하는지 한 번에 확인하는 예제입니다.

import re

txt = "hello heyo heeeo planet The rain in Spain falls and stays 123"

examples = {
    "[a-m]": "[a-m]",          # a부터 m 사이 문자 찾기
    r"\d": r"\d",             # 숫자 찾기
    "he..o": "he..o",         # he + 아무 문자 2개 + o
    "^hello": "^hello",       # hello로 시작하는지 확인
    "123$": "123$",           # 123으로 끝나는지 확인
    "he.*o": "he.*o",         # he로 시작해서 o로 끝나는 가장 긴 형태
    "he.+o": "he.+o",         # he와 o 사이에 문자가 1개 이상 있어야 함
    "he.?o": "he.?o",         # he와 o 사이에 문자가 0개 또는 1개
    "he.{2}o": "he.{2}o",     # he와 o 사이에 정확히 문자 2개
    "falls|stays": "falls|stays"  # falls 또는 stays 찾기
}

for name, pattern in examples.items():
    print(f"{name:12} -> {re.findall(pattern, txt)}")
```

```python
# 4. Flags, 플래그
# Example 3
# 목적: re.IGNORECASE 플래그를 사용하면 대문자/소문자를 무시하고 검색할 수 있음을 보여줍니다.

import re

txt = "Python python PYTHON"

print("플래그 없이 검색:", re.findall("python", txt))
print("re.IGNORECASE 사용:", re.findall("python", txt, re.IGNORECASE))
```

```python
# 5. Special Sequences, 특수 시퀀스
# Example 4
# 목적: \A, \b, \B, \d, \D, \s, \S, \w, \W, \Z 같은 특수 시퀀스를 확인합니다.
# r"" 형태는 raw string으로, 역슬래시를 정규표현식 의미 그대로 사용하게 해줍니다.

import re

txt = "The rain in Spain 2026!"

special_patterns = {
    r"\AThe": r"\AThe",       # 문자열 맨 앞이 The인지
    r"\bain": r"\bain",       # 단어 시작 위치에 ain이 있는지
    r"ain\b": r"ain\b",       # 단어 끝 위치에 ain이 있는지
    r"\Bain": r"\Bain",       # 단어 시작이 아닌 곳의 ain
    r"ain\B": r"ain\B",       # 단어 끝이 아닌 곳의 ain
    r"\d": r"\d",             # 숫자
    r"\D": r"\D",             # 숫자가 아닌 문자
    r"\s": r"\s",             # 공백
    r"\S": r"\S",             # 공백이 아닌 문자
    r"\w": r"\w",             # 문자, 숫자, 밑줄
    r"\W": r"\W",             # 문자, 숫자, 밑줄이 아닌 것
    r"!\Z": r"!\Z"             # 문자열 맨 끝이 !인지
}

for name, pattern in special_patterns.items():
    result = re.findall(pattern, txt)
    print(f"{name:8} -> {result[:10]}")
```

```python
# 6. Sets, 문자 집합
# Example 5
# 목적: [] 문자 집합을 이용해 특정 문자, 숫자 범위, 알파벳 범위를 찾는 예제입니다.

import re

txt = "The rain in Spain + 2026"

set_patterns = {
    "[arn]": "[arn]",            # a, r, n 중 하나
    "[a-n]": "[a-n]",            # a부터 n 사이의 소문자
    "[^arn]": "[^arn]",          # a, r, n이 아닌 문자
    "[0123]": "[0123]",          # 0, 1, 2, 3 중 하나
    "[0-9]": "[0-9]",            # 숫자 0~9
    "[0-5][0-9]": "[0-5][0-9]",  # 00~59 형태의 두 자리 숫자
    "[a-zA-Z]": "[a-zA-Z]",      # 알파벳 대소문자
    "[+]": "[+]"                 # + 문자 자체
}

for name, pattern in set_patterns.items():
    result = re.findall(pattern, txt)
    print(f"{name:10} -> {result[:20]}")
```

```python
# 8. `search()` 함수
# Example 8
# 목적: 문자열에서 첫 번째 공백 문자의 위치를 찾습니다.
# \s는 공백을 의미하고, start()는 찾은 위치의 시작 인덱스를 반환합니다.

import re

txt = "The rain in Spain"
x = re.search(r"\s", txt)
print("The first white-space character is located in position:", x.start())
```

```python
# 11. Match Object
# Example 15
# 목적: span()으로 첫 번째 매칭의 시작 위치와 끝 위치를 확인합니다.
# \bS\w+ 는 S로 시작하는 단어를 찾는 정규표현식입니다.

import re

txt = "The rain in Spain"
x = re.search(r"\bS\w+", txt)
print(x.span())
```

```python
# 13. Python Try Except
# Example 20
# 목적: 정의되지 않은 변수를 출력하면 오류가 나지만,
# try-except를 쓰면 프로그램이 멈추지 않고 오류를 처리할 수 있습니다.

try:
    print(not_defined_x)
except:
    print("An exception occurred")
```

```python
# 13. Python Try Except
# Example 21
# 목적: try 블록이 없다면 NameError가 발생한다는 것을 안전하게 보여줍니다.
# 실제로 print(not_defined_x)를 그대로 실행하면 프로그램이 오류로 멈춥니다.

try:
    eval("not_defined_x")
except NameError as e:
    print("try 없이 실행하면 이런 오류가 발생합니다.")
    print(type(e).__name__ + ":", e)
```

```python
# 13. Python Try Except
# Example 22
# 목적: 오류 종류에 따라 다른 except 블록을 실행하는 예제입니다.
# NameError가 발생하면 첫 번째 except NameError가 실행됩니다.

try:
    print(not_defined_x)
except NameError:
    print("Variable x is not defined")
except:
    print("Something else went wrong")
```

```python
# 13. Python Try Except
# 추가 예제
# 목적: 오류가 없을 때 else가 실행되고, finally는 오류 여부와 상관없이 항상 실행됨을 보여줍니다.

try:
    number = 10 / 2
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
else:
    print("오류가 없어서 else가 실행되었습니다:", number)
finally:
    print("finally는 항상 실행됩니다.")
```

```python
# 예제 설명: 월러스 연산자 :=를 사용해 값을 저장하면서 조건 비교도 하는 예제입니다.
numbers = [1, 2, 3, 4, 5] 
count = len(numbers) 
if count > 3: 
    print(f"List has {count} elements") 
if (count := len(numbers)) > 3: 
    print(f"List has {count} elements")
```

```python
# 예제 설명: 비교 연산자 ==, !=, >, <, >=, <=의 의미를 보여주는 예제입니다.
x = 5 
y = 3 
print(x == y) 
print(x != y) 
print(x > y) 
print(x < y) 
print(x >= y) 
print(x <= y)
```

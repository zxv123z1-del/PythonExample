```python
# 예제 설명: 딕셔너리는 key와 value를 한 쌍으로 저장하는 자료형이라는 것을 보여줍니다.
thisdict = { 
    "brand": "Ford", 
    "model": "Mustang", 
    "year": 1964 
} 
print(thisdict)
```

```python
# 예제 설명: 딕셔너리에서 key를 사용해 원하는 value를 꺼내는 예제입니다.
thisdict = { 
    "brand": "Ford", 
    "model": "Mustang",    
    "year": 1964 
} 
print(thisdict["brand"])
```

```python
# 예제 설명: 같은 key가 두 번 나오면 뒤의 값이 앞의 값을 덮어쓰며, len()은 key 개수를 센다는 예제입니다.
thisdict = { 
    "brand": "Ford", 
    "model": "Mustang", 
    "year": 1964, 
    "year": 2020 
} 
print(thisdict)

print(len(thisdict))
```

```python
# 예제 설명: 딕셔너리 value에는 문자열, 숫자, 불리언, 리스트 등 다양한 자료형을 넣을 수 있다는 예제입니다.
thisdict = { 
"brand": "Ford", 
"electric": False, 
"year": 1964, 
"colors": ["red", "white", "blue"] 
}
```

```python
# 예제 설명: type()으로 딕셔너리 자료형이 dict인지 확인하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
print(type(thisdict))
```

```python
# 예제 설명: 대괄호와 key를 사용해 딕셔너리 값을 가져오는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
x = thisdict["model"]
```

```python
# 예제 설명: get()으로 딕셔너리 값을 안전하게 가져오는 예제입니다.
x = thisdict.get("model")
```

```python
# 예제 설명: keys()로 딕셔너리의 key 목록을 가져오는 예제입니다.
x = thisdict.keys()
```

```python
# 예제 설명: keys()로 얻은 view는 딕셔너리에 key가 추가되면 같이 바뀐다는 예제입니다.
car = { 
  "brand": "Ford", 
  "model": "Mustang", 
  "year": 1964 
} 
x = car.keys() 
print(x) #before the change 
car["color"] = "white" 
print(x) #after the change
```

```python
# 예제 설명: values()로 얻은 view는 딕셔너리 값이 추가되면 같이 바뀐다는 예제입니다.
car = { 
  "brand": "Ford", 
  "model": "Mustang", 
  "year": 1964 
} 
x = car.values() 
print(x) #before the change 
car["color"] = "red" 
print(x) #after the change
```

```python
# 예제 설명: items()는 key와 value 쌍을 보여주며, 값이 바뀌면 결과도 바뀐다는 예제입니다.
car = { 
  "brand": "Ford", 
  "model": "Mustang", 
  "year": 1964 
} 
x = car.items() 
print(x) #before the change 
car["year"] = 2020 
print(x) #after the change
```

```python
# 예제 설명: items()는 새 key-value가 추가되어도 자동으로 반영된다는 예제입니다.
car = { 
  "brand": "Ford", 
  "model": "Mustang", 
  "year": 1964 
} 
x = car.items() 
print(x) #before the change 
car["color"] = "red" 
print(x) #after the change
```

```python
# 예제 설명: in을 사용해 특정 key가 딕셔너리에 있는지 확인하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
if "model" in thisdict: 
    print("Yes, 'model' is one of the keys in the thisdict dictionary")
```

```python
# 예제 설명: 기존 key에 새 값을 대입하면 value가 수정된다는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
thisdict["year"] = 2018
```

```python
# 예제 설명: update()로 기존 key의 값을 수정하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
thisdict.update({"year": 2020})
```

```python
# 예제 설명: 새 key에 값을 대입하면 딕셔너리에 항목이 추가된다는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
thisdict["color"] = "red" 
print(thisdict)
```

```python
# 예제 설명: update()로 새 key-value 항목을 추가하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
thisdict.update({"color": "red"})
```

```python
# 예제 설명: pop(key)으로 특정 key와 value를 삭제하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
thisdict.pop("model") 
print(thisdict)
```

```python
# 예제 설명: popitem()은 마지막으로 추가된 항목을 삭제하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
thisdict.popitem() 
print(thisdict)
```

```python
# 예제 설명: del을 사용해 특정 key-value 항목을 삭제하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
del thisdict["model"] 
print(thisdict)
```

```python
# 예제 설명: del로 딕셔너리 변수 자체를 지우면 이후 사용할 때 오류가 난다는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
del thisdict 
print(thisdict) 
#this will cause an error because "thisdict" no longer exists.
```

```python
# 예제 설명: clear()는 딕셔너리는 남겨두고 내부 항목만 모두 비운다는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
thisdict.clear() 
print(thisdict)
```

```python
# 예제 설명: for문으로 딕셔너리를 반복하면 기본적으로 key가 나온다는 예제입니다.
for x in thisdict: 
    print(x)
```

```python
# 예제 설명: key를 이용해 반복하면서 각 key의 value를 출력하는 예제입니다.
for x in thisdict: 
    print(thisdict[x])
```

```python
# 예제 설명: values()를 사용해 value만 반복 출력하는 예제입니다.
for x in thisdict.values(): 
    print(x)
```

```python
# 예제 설명: keys()를 사용해 key만 반복 출력하는 예제입니다.
for x in thisdict.keys(): 
    print(x)
```

```python
# 예제 설명: items()를 사용해 key와 value를 함께 반복 출력하는 예제입니다.
for x,y in thisdict.items(): 
    print(x, y)
```

```python
# 예제 설명: copy()로 딕셔너리를 복사하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
mydict = thisdict.copy() 
print(mydict)
```

```python
# 예제 설명: dict() 생성자를 사용해 딕셔너리를 복사하는 예제입니다.
thisdict = { 
"brand": "Ford", 
"model": "Mustang", 
"year": 1964 
} 
mydict = dict(thisdict) 
print(mydict)
```

```python
# 예제 설명: 딕셔너리 안에 또 다른 딕셔너리를 넣는 중첩 딕셔너리 예제입니다.
myfamily = { 
"child1" : { 
"name" : "Emil", 
"year" : 2004 
  }, 
"child2" : { 
"name" : "Tobias", 
"year" : 2007 
  }, 
"child3" : { 
"name" : "Linus", 
"year" : 2011 
  } 
}
```

```python
# 예제 설명: 각 자식 딕셔너리를 따로 만든 뒤 하나의 딕셔너리로 묶는 예제입니다.
child1 = { 
"name" : "Emil", 
"year" : 2004 
} 
child2 = { 
"name" : "Tobias", 
"year" : 2007 
} 
child3 = { 
"name" : "Linus", 
"year" : 2011 
} 
myfamily = { 
"child1" : child1, 
"child2" : child2, 
"child3" : child3
}
```

```python
# 예제 설명: 중첩 딕셔너리에서 바깥 key와 안쪽 key를 순서대로 사용해 값을 꺼내는 예제입니다.
print(myfamily["child2"]["name"])
```

```python
# 예제 설명: 중첩 딕셔너리를 이중 for문으로 돌며 모든 값을 출력하는 예제입니다.
for x, obj in myfamily.items(): 
    print(x) 
    for y in obj: 
        print(y + ':', obj[y])
```
